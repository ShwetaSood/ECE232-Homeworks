```python
import json
import pandas as pd
with open("/Users/sonali/Desktop/EE232/Project5/Data/san_francisco_censustracts.json") as f:
    data = json.load(f)
```

```python
from pandas.io.json import json_normalize
df = json_normalize(data["features"])
```

```python
df.head()
```

```python
location=df[['properties.MOVEMENT_ID','properties.DISPLAY_NAME']]
location.to_csv("location.csv", encoding='utf-8', index=False)
```

```python
count =0
coor=open("/Users/sonali/Desktop/EE232/Project5/Data/coordinates.txt", 'w')
for i in df['properties.MOVEMENT_ID']:
#     print(type(i))
    sub=df[df['properties.MOVEMENT_ID']==i]
    x=sub['geometry.coordinates'][count][0][0]
#     print x
    lat=0
    lon=0
    c=0
    for j in x:
        
        lat+=j[0]
        lon+=j[1]
        c+=1
    lat=float(lat)/c
    lon=float(lon)/c
#     print lat, lon
    coor.write(str(i)+'\t'+str(lat)+'\t'+str(lon))
    coor.write('\n')
    count=count+1
#     print x
#     break
# x=df['geometry.coordinates'][:][0][0]
# x
coor.close()
```
