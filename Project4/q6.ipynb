```python
import re
actor_read_file = open("data/actor_movies.txt", 'r')
actress_read_file = open("data/actress_movies.txt", 'r')

# process_file = open("data/processed_actors_actresses.txt", 'w')

threshold = 10
id = 0
cnt=0
d={}
unique_movies=set()

for line in actor_read_file.readlines():
    line = line[:-1]
    line = re.sub("^\\(", "",line)
    line = re.sub("\\([^[:digit:]]+\\)", "",line)
    line = re.sub("^\\s+|\\s+$", "",line)
    temp = line.split("\t\t")
    temp=[t.strip() for t in temp]
    if len(temp)< threshold+1:
        continue
    temp[0].strip(" ")
    temp[0].strip("\t")
    
    mv=temp[1:]
    for i in range(0,len(mv)):
        m=mv[i].split(")")
        mv[i]=m[0]+")"
    mv=set(mv)
    unique_movies.update(mv)
    d[id]=mv
    id +=1
    
    #break
    #process_file.write(str(temp[0]) + "\t\t" + str(id)) 
    #process_file.write("\n")
    cnt = cnt+1

print("Finished actors")

for line in actress_read_file.readlines():
    #break
    line = line[:-1]
    line = re.sub("^\\(", "",line)
    line = re.sub("\\([^[:digit:]]+\\)", "",line)
    line = re.sub("^\\s+|\\s+$", "",line)
    temp=line.split("\t\t")
    temp=[t.strip() for t in temp]
    if len(temp)< threshold+1:
        continue
    temp[0].strip(" ")
    temp[0].strip("\t")
    
   
    mv=temp[1:]
    for i in range(0,len(mv)):
        m=mv[i].split(")")
        mv[i]=m[0]+")"
    mv=set(mv)
    unique_movies.update(mv)
    d[id]=mv
    id +=1
    
    #process_file.write(str(temp[0]) + "\t\t" + str(id))
    #process_file.write("\n")
    cnt = cnt+1

print("Finished actresses")

print("Total number of actors and actresses:" , cnt)
print("Total unique movies:" , len(unique_movies))
actor_read_file.close()
actress_read_file.close()
#process_file.close()
```

```python
um = open("data/unique_movies.txt", 'w')
for u in unique_movies:
    um.write(u)
    um.write('\n')
um.close()
```
